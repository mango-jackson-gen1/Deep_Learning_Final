# FMA Dataset - Exploratory Data Analysis

Quick overview of the FMA small subset: genre distribution, track durations, and artist coverage.

```python
import sys
sys.path.insert(0, '..')

import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
from src.metadata import load_tracks, load_genres, get_small_subset_ids

sns.set_style('whitegrid')
%matplotlib inline
```

```python
tracks = load_tracks()
genres = load_genres()
small_ids = get_small_subset_ids(tracks)
small = tracks.loc[small_ids]
print(f'Small subset: {len(small)} tracks, {len(genres)} genres')
```

## Genre Distribution

```python
top_genre = small[('track', 'genre_top')]
genre_counts = top_genre.value_counts()

fig, ax = plt.subplots(figsize=(10, 5))
genre_counts.plot.barh(ax=ax)
ax.set_xlabel('Number of Tracks')
ax.set_title('Genre Distribution (FMA Small)')
plt.tight_layout()
plt.show()
print(genre_counts)
```

## Track Duration Distribution

```python
durations = small[('track', 'duration')].dropna()

fig, ax = plt.subplots(figsize=(10, 4))
durations.clip(upper=600).hist(bins=50, ax=ax, edgecolor='black')
ax.set_xlabel('Duration (seconds, clipped at 600s)')
ax.set_ylabel('Count')
ax.set_title('Track Duration Distribution')
plt.tight_layout()
plt.show()
print(durations.describe())
```

## Artist Coverage

```python
if ('artist', 'name') in small.columns:
    n_artists = small[('artist', 'name')].nunique()
    tracks_per_artist = small[('artist', 'name')].value_counts()
    print(f'Unique artists: {n_artists}')
    print(f'\nTop 15 artists by track count:')
    print(tracks_per_artist.head(15))
else:
    print('Artist name column not found in this metadata format')
```
