# Echo Nest Feature Exploration

This notebook explores the Echo Nest audio features (valence, energy, etc.) and shows how they can be used as filters for semantic retrieval.

```python
import os
import sys
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Add project root to path
sys.path.append(os.path.abspath(".."))

from src.config import FMA_METADATA_DIR
from src.metadata import load_tracks, get_small_subset_ids
```

### Load Data

```python
tracks = load_tracks(FMA_METADATA_DIR)
small_ids = get_small_subset_ids(tracks)
tracks_small = tracks.loc[small_ids]

# Load echonest
echonest = pd.read_csv(FMA_METADATA_DIR / "echonest.csv", index_col=0, header=[0, 1, 2])
print(f"Loaded Echo Nest features for {len(echonest)} tracks.")

# Merge with small subset
merged = pd.merge(tracks_small[[]], echonest['echonest']['audio_features'], left_index=True, right_index=True, how='inner')
print(f"Overlap: {len(merged)} tracks in 'small' subset have Echo Nest features.")
```

### Feature Distribution

```python
plt.figure(figsize=(12, 8))
features = ['valence', 'energy', 'danceability', 'acousticness']
for i, feat in enumerate(features):
    plt.subplot(2, 2, i+1)
    sns.histplot(merged[feat], kde=True)
    plt.title(f'Distribution of {feat}')
plt.tight_layout()
plt.show()
```

### High Energy vs Low Valence Filtering

```python
# Example: find 'Aggressive' tracks (high energy, low valence)
aggressive = merged[(merged['energy'] > 0.8) & (merged['valence'] < 0.3)]
print(f"Found {len(aggressive)} aggressive tracks.")
print(aggressive.head())
```
